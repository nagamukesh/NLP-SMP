16. Write a Pyorch snippet to create an nxn(input to function) tensor with random values. Create an array from the said array swapping first and last rows.

````python
def swaping(n):
  tsr = torch.rand(n,n)
  tsr_clone = tsr.clone()
  tsr[:,0] = tsr_clone[:,n-1]
  tsr[:,n-1] = tsr_clone[:,0]
  return tsr, tsr_clone

swaping(5)
````

1. Given a 3D tensor of shape `(a, b, c)`*italicised text*, permute its axes to obtain tensors of all other shape combinations (i.e 5).

````python
def permute(a,b,c):
  tensor_3d = torch.rand(a, b, c)

  tensor_acb = tensor_3d.permute(0, 2, 1)  # Shape (a, c, b)
  tensor_bac = tensor_3d.permute(1, 0, 2)  # Shape (b, a, c)
  tensor_bca = tensor_3d.permute(1, 2, 0)  # Shape (b, c, a)
  tensor_cab = tensor_3d.permute(2, 0, 1)  # Shape (c, a, b)
  tensor_cba = tensor_3d.permute(2, 1, 0)  # Shape (c, b, a)

  print("Original tensor shape (a, b, c):", tensor_3d.shape)
  print("Permuted tensor shape (a, c, b):", tensor_acb.shape)
  print("Permuted tensor shape (b, a, c):", tensor_bac.shape)
  print("Permuted tensor shape (b, c, a):", tensor_bca.shape)
  print("Permuted tensor shape (c, a, b):", tensor_cab.shape)
  print("Permuted tensor shape (c, b, a):", tensor_cba.shape)

  return

permute(2,4,6)
````

2. One of the most important aspects of improving model performance ins normalization. Implement a custom NN normalization layer that performs either a **batch** or **layer** normalization (based on a **dim** parameter)and integrate it into a simple feedforward neural network.
- Note that you dont need to implement any backward() here, for simplicity.
- Consider your input tensor to the neural network as of shape (m, n) where m is num of examples, n is num of features.
- Which value of **dim** leads to batch and layer normalization?

````python
m = 8 # num of examples
n = 5 # num of features
inp_tsr = torch.randint(10,20,(m,n)).float()
inp_tsr
````

````python
import torch.nn as nn

class CustomNormLayer(nn.Module):
  def __init__(self, num_features, dim:str):
    super().__init__()

    if dim == 'batch':
      self.norm = nn.BatchNorm1d(num_features)
    elif dim == 'layer':
      self.norm = nn.LayerNorm(num_features)

  def forward(self,tensor):
    return self.norm(tensor)
````

````python
inp_tsr = inp_tsr
batch_norm = CustomNormLayer(num_features = inp_tsr.size(1), dim ='batch')
layer_norm = CustomNormLayer(num_features = inp_tsr.size(1), dim ='layer')
batch_norm(inp_tsr), layer_norm(inp_tsr)
````
